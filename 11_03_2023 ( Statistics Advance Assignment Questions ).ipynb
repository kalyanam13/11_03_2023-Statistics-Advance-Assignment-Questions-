### Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.
### Answer : The t-test and z-test are both statistical tests used to make inferences about population parameters based on sample data, but they are applicable in different situations.

### T-test:
### The t-test is typically used when the sample size is small (less than 30) or when the population standard deviation is unknown. It is particularly suitable for situations where you are comparing the means of two groups.

### Example scenario for a t-test:
### Suppose you want to test whether there is a significant difference in the average test scores of two different teaching methods. You collect a sample of 20 students who were taught using Method A and another sample of 20 students taught using Method B. You can use a t-test to determine if there is a statistically significant difference in the mean scores between the two groups.

### Z-test:
### The z-test is used when the sample size is large (typically n > 30) and the population standard deviation is known. It is commonly employed when testing hypotheses about the population mean.

### Example scenario for a z-test:
### Imagine you have access to the test scores of 500 students in a population, and you want to test if the average score is significantly different from a hypothesized value. If you know the population standard deviation, you can use a z-test to assess whether the mean test score is significantly different from the hypothesized value.

### In summary, the choice between a t-test and a z-test depends on factors such as the sample size and whether the population standard deviation is known. Use a t-test for small sample sizes or when the population standard deviation is unknown, and use a z-test for larger sample sizes with a known population standard deviation.

### Q2: Differentiate between one-tailed and two-tailed tests.
### Answer : 

One-tailed and two-tailed tests refer to the directionality of the statistical hypothesis and the corresponding critical regions in hypothesis testing.

One-tailed test:
In a one-tailed test, the null hypothesis specifies a particular direction (greater than or less than), and the alternative hypothesis focuses on that specific direction. The critical region is located on only one side of the distribution curve. One-tailed tests are often used when there is a specific expectation or prediction about the direction of the effect.

Example:
Null hypothesis (H0): The average score is equal to 50.
Alternative hypothesis (H1): The average score is greater than 50.
Critical region: Located in the right tail of the distribution.
Two-tailed test:
In a two-tailed test, the null hypothesis does not make a specific directional prediction, and the alternative hypothesis simply states that there is a difference. The critical region is divided into two sides of the distribution curve. Two-tailed tests are used when you are interested in determining if there is a significant difference in either direction.

Example:
Null hypothesis (H0): The average scores are equal between two groups.
Alternative hypothesis (H1): The average scores are different between two groups.
Critical region: Located in both the tails of the distribution.
The choice between a one-tailed and a two-tailed test depends on the nature of the research question and the specific hypothesis you want to test. One-tailed tests may be appropriate when there is a clear directional expectation, while two-tailed tests are more conservative and are used when you are interested in detecting differences in either direction.

### Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.
### Answer : 

Type I Error:
A Type I error occurs when the null hypothesis is incorrectly rejected when it is actually true. In other words, it's a false positive or a conclusion that there is a significant effect or difference when there isn't one. The probability of making a Type I error is denoted by the significance level (α), which is set by the researcher.

Example Scenario for Type I Error:
Suppose a pharmaceutical company is testing a new drug, and the null hypothesis (H0) is that the drug has no effect (i.e., it is equivalent to a placebo). The significance level (α) is set at 0.05. If, based on the sample data, the researchers reject the null hypothesis and conclude that the drug is effective when, in reality, it has no effect, a Type I error has occurred.

Type II Error:
A Type II error occurs when the null hypothesis is incorrectly not rejected when it is actually false. In other words, it's a false negative or a failure to detect a real effect or difference that exists. The probability of making a Type II error is denoted by the symbol β.

Example Scenario for Type II Error:
Continuing with the drug example, let's say the drug does have a real effect, but the study fails to detect it. The null hypothesis (H0) is not rejected, and the researchers conclude that the drug is not effective when it actually is. In this case, a Type II error has occurred.

The relationship between Type I and Type II errors is often described by the concept of statistical power. Statistical power is the probability of correctly rejecting a false null hypothesis (1 - β). Researchers aim to strike a balance between minimizing Type I and Type II errors based on the specific goals and risks associated with a study.
